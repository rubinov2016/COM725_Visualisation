```python
import pandas as pd

dataVal = [(10,20,30,40,50,60,70),
          (10,10,40,40,50,60,70),
          (10,20,30,50,50,60,80)]

dataFrame =  pd.DataFrame(data=dataVal)
skewValue = dataFrame.skew(axis=1)

print("DataFrmae:")
print(dataFrame)

print("Skew:")
print(skewValue)


kurtValue = dataFrame.kurtosis(axis=1)
print("Kurt:")
print(kurtValue)
```

```python
dataFrame.plot(kind='bar')
```
